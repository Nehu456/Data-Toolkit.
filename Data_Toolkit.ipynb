NumPy (Numerical Python) is a powerful open-source library for numerical computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on them efficiently.

Why is NumPy Widely Used?

**Performance** – NumPy operations are significantly faster than native Python lists due to its use of optimized C and Fortran code.

**Multidimensional Arrays** – The ndarray object allows efficient storage and manipulation of large datasets.


**Broadcasting **– Enables element-wise operations on arrays of different shapes without explicit looping.

Answer 2

The groupby() method in Pandas is used to group data based on one or more columns and then apply aggregation or transformation functions to analyze the grouped data efficiently.

It follows a split-apply-combine approach:

Split – Divides the dataset into groups based on a column.


Apply – Performs an operation (e.g., sum, mean, count) on each group.


Combine – Merges the results back into a DataFrame.

A heatmap is a type of data visualization that represents values in a matrix or table format using colors. The intensity of the color corresponds to the magnitude of the data points, making it easy to identify patterns, correlations, and outliers in large datasets.


When Should a Heatmap Be Used?

✅ 1. Analyzing Correlations (Correlation Matrix)

Used to visualize relationships between multiple variables in a dataset.
Example: Finding correlations between stock prices, sales, or medical conditions.


✅ 2. Identifying Trends & Patterns

Helps spot trends over time or across categories.
Example: Website traffic heatmaps show peak hours of user activity.


✅ 3. Highlighting Anomalies & Outliers

Useful for detecting unusual values in datasets.
Example: Fraud detection in financial transactions.

```python
#answer 4

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load sample dataset
df = sns.load_dataset("penguins").dropna()  # Remove NaN values

# Select only numerical columns
numeric_df = df.select_dtypes(include=["number"])


corr_matrix = numeric_df.corr()
print(corr_matrix)

# Create a heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Add a title
plt.title("Correlation Matrix Heatmap")

# Show plot
plt.show()

```

```python
#answer 8

import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)  # Generates 100 points between 0 and 10
y1 = np.sin(x)  # First line (sine wave)
y2 = np.cos(x)  # Second line (cosine wave)
y3 = np.tan(x)  # Third line (tangent)

# Create the plot
plt.figure(figsize=(8, 5))

# Plot multiple lines
plt.plot(x, y1, label="Sine Wave", color="blue", linestyle="-", linewidth=2)
plt.plot(x, y2, label="Cosine Wave", color="red", linestyle="--", linewidth=2)
plt.plot(x, y3, label="Tangent", color="green", linestyle=":", linewidth=2)

# Add labels, title, and legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot Example")
plt.legend()

# Show the plot
plt.show()
```

```python
#answer 13

import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(42)
x = np.random.randn(50)
y = np.random.randn(50)
z = np.random.randn(50)

# Create 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x, y=y, z=z,
    mode='markers',
    marker=dict(
        size=6,
        color=z,  # Color by z values
        colorscale='Viridis',
        opacity=0.8
    )
)])

# Add title & labels
fig.update_layout(
    title="3D Scatter Plot with Plotly",
    scene=dict(
        xaxis_title="X Axis",
        yaxis_title="Y Axis",
        zaxis_title="Z Axis"
    )
)

# Show plot
fig.show()
```
